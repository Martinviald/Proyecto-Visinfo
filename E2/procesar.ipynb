```python
# Importamos el data set de terremotos delimitado por ;
import pandas as pd

df = pd.read_csv('chile_earthquakes.csv', delimiter=';')
df = df.drop(columns=['Search Parameters'])
df = df.fillna(0)
# Cambiamos el nombre Location Name por LocationName, Focal Depth (km) por FocalDepth, Damage ($Mil) por Damage, Houses Destroyed por HousesDestroyed, Houses Damaged por HousesDamaged, Total Deaths por TotalDeaths, Total Injuries por TotalInjuries, Total Damage ($Mil) por TotalDamage, Total Houses Destroyed por TotalHousesDestroyed, Total Houses Damaged por TotalHousesDamaged y Houses Damaged por HousesDamaged.
df = df.rename(columns={'Location Name': 'LocationName', 'Focal Depth (km)': 'FocalDepth', 'Damage ($Mil)': 'Damage', 'Houses Destroyed': 'HousesDestroyed', 'Houses Damaged': 'HousesDamaged', 'Total Deaths': 'TotalDeaths', 'Total Injuries': 'TotalInjuries', 'Total Damage ($Mil)': 'TotalDamage', 'Total Houses Destroyed': 'TotalHousesDestroyed', 'Total Houses Damaged': 'TotalHousesDamaged', 'Houses Damaged': 'HousesDamaged'})
# Eliminamos el primer dato
df = df.drop(df.index[0])
df = df[df.Magnitude != 0]
# Eliminmos las columnas que tienen cero en la columna FocalDepth
df = df[df.FocalDepth != 0]
# Agregamos una columna id y la ubicamos de primera
df['id'] = range(1, 1+len(df))
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
df
```

```python
# Convertimos todos los datos a strings
df = df.astype(str)
```

```python
df.columns
```

```python
# Reemplazamos todos los valores que tengan "," por "." de las columnas ['id', 'Year', 'Month', 'Day', 'Hour', 'Minute', 'Tsunami', 'Country', 'LocationName', 'Latitude', 'Longitude', 'FocalDepth', 'Magnitude', 'Deaths', 'Missing', 'Injuries', 'Damage', 'HousesDestroyed', 'HousesDamaged']
df = df.replace({',': '.'}, regex=True)
```

```python
# convertimos todos los datos de las columnas ['id', 'Year', 'Month', 'Day', 'Hour', 'Minute', 'Tsunami', 'Country', 'Latitude', 'Longitude', 'FocalDepth', 'Magnitude', 'Deaths', 'Missing', 'Injuries', 'Damage', 'HousesDestroyed', 'HousesDamaged'] a float
df[['id', 'Year', 'Month', 'Day', 'Hour', 'Minute', 'Tsunami', 'Latitude', 'Longitude', 'FocalDepth', 'Magnitude', 'Deaths', 'Missing', 'Injuries', 'Damage', 'HousesDestroyed', 'HousesDamaged']] = df[['id', 'Year', 'Month', 'Day', 'Hour', 'Minute', 'Tsunami', 'Latitude', 'Longitude', 'FocalDepth', 'Magnitude', 'Deaths', 'Missing', 'Injuries', 'Damage', 'HousesDestroyed', 'HousesDamaged']].astype(float)
```
